### Defining IOU metric and compile Model

```python
from keras.losses import binary_crossentropy
from keras import backend as K
import tensorflow as tf

def dice_loss(y_true, y_pred):
    y_true = tf.cast(y_true, dtype=tf.float32)
    y_pred = tf.cast(y_pred, dtype=tf.float32)
    smooth = 1.
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = y_true_f * y_pred_f
    score = (2. * K.sum(intersection) + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)
    return 1. - score

### bce_dice_loss = binary_crossentropy_loss + dice_loss
def bce_dice_loss(y_true, y_pred):
    return binary_crossentropy(y_true, y_pred) + dice_loss(y_true, y_pred)
```

```python
from keras.layers import Conv2D, Activation, Add, MaxPooling2D, Conv2DTranspose, concatenate, Dropout, Input, BatchNormalization, UpSampling2D, PReLU, Dense, Lambda
from keras.models import Model
```

```python
class W_T_E_net(tf.keras.Model):
    def __init__(self) -> None:
        super(W_T_E_net, self).__init__()
        # self.W_T = self.build_W_T(input_shape_W_T)
        # self.E = self.build_E(input_shape_E)

    def conv_block(self, input, num_filters, dilation_rate = 1):
        x = Conv2D(num_filters, (3,3), padding="same", dilation_rate=dilation_rate)(input)
        x = BatchNormalization()(x)
        x = PReLU()(x)
        return x
    
    def conv_block2(self, input, num_filters, dilation_rate = 1):
        x = Conv2D(num_filters, (1,1), padding="same", dilation_rate=dilation_rate)(input)
        x = BatchNormalization()(x)
        x = PReLU()(x)
        return x
    
    def residual_block(self, input, num_filters, dilation_rate = 1):
        x = self.conv_block(input, num_filters, dilation_rate)
        x = self.conv_block(x, num_filters)
        x = Add()([x, input])
        return x
    # def croped(self, datas, sizes, centers, croped_size):
    #     min_coord = []
    #     max_coord = []
    #     for i, center in enumerate(centers):
    #         data = datas[i, :, :]
    #         bounding = tf.cast(data.shape, tf.float64)
    #         diff_max = bounding - (center + tf.constant([croped_size/2, croped_size/2], tf.float64))
    #         diff_max = tf.where(tf.less(diff_max, 0), diff_max, 0)
    #         center = center + diff_max
    #         diff_min = center - tf.constant([croped_size/2, croped_size/2], tf.float64)
    #         diff_min = tf.where(tf.less(diff_min, 0), diff_min, 0)
    #         center = center - diff_min
    #         min_coord.append(tf.cast(center - tf.constant([croped_size/2, croped_size/2], tf.float64)),tf.int32)
    #         max_coord.append(tf.cast(center + tf.constant([croped_size/2, croped_size/2], tf.float64)),tf.int32)

        # return min_coord, max_coord
    def handle_input_E(self, inputs, get_min_coord=False):
        masks, inputs = inputs
        croped_inputs = []
        store_min_coords = []
        croped_size = 64
        masks = tf.squeeze(masks, axis=-1) 
        masks = tf.unstack(masks, axis=0, num=32)
        inputs = tf.squeeze(inputs, axis=-1)
        inputs = tf.unstack(inputs, axis=0, num=32)
        for mask, image in zip(masks, inputs):
            croped_image = tf.zeros((croped_size, croped_size))
            nonzero_coords = tf.where(tf.greater_equal(mask, 0.5))
            min_coord = tf.reduce_min(nonzero_coords, axis=0)
            max_coord = tf.reduce_max(nonzero_coords, axis=0)
            size = max_coord - min_coord
            center = tf.math.round((min_coord + max_coord)/2)

            data = image
            bounding = tf.cast(data.shape, tf.float64) - tf.constant([1, 1], tf.float64)
            diff_max = bounding - (center + tf.constant([croped_size/2, croped_size/2], tf.float64))
            diff_max = tf.where(tf.less(diff_max, 0), diff_max, 0)
            center = center + diff_max
            diff_min = center - tf.constant([croped_size/2, croped_size/2], tf.float64)
            diff_min = tf.where(tf.less(diff_min, 0), diff_min, 0)
            center = center - diff_min
            min_coords = tf.cast(center - tf.constant([croped_size/2, croped_size/2], tf.float64), tf.int32)
            max_coords = tf.cast(center + tf.constant([croped_size/2, croped_size/2], tf.float64), tf.int32)
            croped_image += image[min_coords[0]:max_coords[0], min_coords[1]:max_coords[1]]
            croped_image = tf.expand_dims(croped_image, -1)
            croped_inputs.append(croped_image)
            store_min_coords.append(min_coords)
        if get_min_coord:
            return tf.stack(store_min_coords)
        else:
            return tf.stack(croped_inputs)
    
    def update_array(self, inputs):
        original_arrays, replacement_arrays, start_coords = inputs
        result = []
        original_arrays = tf.squeeze(original_arrays, axis=-1)
        original_arrays = tf.unstack(original_arrays, axis=0, num=32)
        replacement_arrays = tf.squeeze(replacement_arrays, axis=-1)
        replacement_arrays = tf.unstack(replacement_arrays, axis=0, num=32)
        start_coords = tf.unstack(start_coords, axis=0, num=32)
        # shape = original_arrays.shape
        for original_array, replacement_array, start_coord in zip(original_arrays, replacement_arrays, start_coords):
            start_row = start_coord[0]
            start_col = start_coord[1]
            # Determine the shape of the replacement array
            replacement_shape = tf.shape(replacement_array)
            indices_ = tf.where(tf.ones(replacement_shape))
            # Replace the part of the original array with the replacement array
            try:
                replaced_array = tf.tensor_scatter_nd_update(original_array,
                                                            indices=indices_ + [[start_row, start_col]],
                                                            updates=tf.reshape(replacement_array, [-1]))
                result.append(tf.expand_dims(replaced_array, -1))
            except Exception as e:
                print(f"Error at {i}")   
                print(f"Original array shape: {original_array.shape}")
                print(f"Replacement array shape: {replacement_array.shape}")
                print(f"Start coord: {start_coord}")        
                raise e
                
        
        return tf.stack(result)

    def build_W_T(self, input_shape):
        # W_T
        input = Input(input_shape, batch_size=32) # 128, 128, 1
        x = self.residual_block(input, 32) # 128, 128, 32
        x = self.residual_block(x, 32)  # 128, 128, 32
        x = self.conv_block2(x, 32) # 128, 128, 32
        x = MaxPooling2D((2,2))(x) # 64, 64, 32
        x = self.residual_block(x, 32) # 64, 64, 32
        x = self.residual_block(x, 32) # 64, 64, 32
        x = self.conv_block2(x, 32) # 64, 64, 32
        x2 = Conv2D(2, (3,3), padding="same") (x) # 64, 64, 2
        x2 = UpSampling2D((2,2), interpolation="bilinear")(x2) # 128, 128, 2
        x = MaxPooling2D((2,2))(x) # 32, 32, 32
        x = self.residual_block(x, 32)
        x = self.residual_block(x, 32, 2)
        x = self.residual_block(x, 32, 3)
        x = self.conv_block2(x, 32) # 32, 32, 32
        x3 = Conv2D(2, (3,3), padding="same") (x) # 128, 128, 2
        x3 = UpSampling2D((4,4), interpolation="bilinear")(x3) # 128, 128, 2
        x = self.residual_block(x, 32, 3)
        x = self.residual_block(x, 32, 2)
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32) # 32, 32, 32
        x4 = Conv2D(2, (3,3), padding="same") (x) # 128, 128, 2
        x4 = UpSampling2D((4,4), interpolation="bilinear")(x4) # 128, 128, 2
        x4 = Add()([x2, x3, x4]) # 128, 128, 2
        concat = concatenate([x4, input]) # 128, 128, 8
        WT_output = Conv2D(2, (3,3), padding="same")(concat) # 128, 128, 2
        WT_output = Dense(1, activation='sigmoid')(WT_output) # 128, 128, 1

        E_input = Lambda(self.handle_input_E, output_shape=(64,64,1))([WT_output, input])
        min_coords = Lambda(self.handle_input_E, arguments={'get_min_coord': True})([WT_output, input])

        x = self.residual_block(E_input, 32) 
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32)
        x = self.residual_block(x, 32)
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32)
        x1 = Conv2D(2, (3,3), padding="same") (x) # 64, 64, 2
        x = MaxPooling2D((2,2))(x)
        x = self.residual_block(x, 32)
        x = self.residual_block(x, 32, 2)
        x = self.residual_block(x, 32, 3)
        x = self.conv_block2(x, 32)
        x2 = Conv2D(2, (3,3), padding="same") (x)
        x2 = UpSampling2D((2,2), interpolation="bilinear")(x2) # 64, 64, 2
        x = self.residual_block(x, 32, 3)
        x = self.residual_block(x, 32, 2)
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32)
        x3 = Conv2D(2, (3,3), padding="same") (x)
        x3 = UpSampling2D((2,2), interpolation="bilinear")(x3) # 64, 64, 2
        x3 = Add()([x1, x2, x3]) # 64, 64, 2
        concat = concatenate([x3, E_input]) # 64, 64, 3
        output = Conv2D(2, (3,3), padding="same")(concat) # 64, 64, 2
        output = Conv2D(1, (3,3), padding="same")(output) # 64, 64, 1
        output = Lambda(self.update_array)([WT_output, output, min_coords])

        output = Dense(1, activation='sigmoid')(output) 
        
        return Model(inputs=[input], outputs=[output])
    def build_E(self, input_shape):
      # W_T
        input = Input(input_shape, batch_size=32) # 64, 64, 1
        x = self.residual_block(input, 32)
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32)
        x = self.residual_block(x, 32)
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32)
        x1 = Conv2D(2, (3,3), padding="same") (x) # 64, 64, 2
        x = MaxPooling2D((2,2))(x)
        x = self.residual_block(x, 32)
        x = self.residual_block(x, 32, 2)
        x = self.residual_block(x, 32, 3)
        x = self.conv_block2(x, 32)
        x2 = Conv2D(2, (3,3), padding="same") (x)
        x2 = UpSampling2D((2,2), interpolation="bilinear")(x2) # 64, 64, 2
        x = self.residual_block(x, 32, 3)
        x = self.residual_block(x, 32, 2)
        x = self.residual_block(x, 32)
        x = self.conv_block2(x, 32)
        x3 = Conv2D(2, (3,3), padding="same") (x)
        x3 = UpSampling2D((2,2), interpolation="bilinear")(x3) # 64, 64, 2
        x3 = Add()([x1, x2, x3]) # 64, 64, 2
        concat = concatenate([x3, input]) # 64, 64, 3
        output = Conv2D(2, (3,3), padding="same")(concat)
        output = Conv2D(1, (3,3), padding="same")(output)
        output = Dense(1, activation='sigmoid')(output) 
        return Model(inputs=[input], outputs=[output])
    
    def call(self, inputs, training=True, mask=None):
        try:
            WT_output = self.W_T((inputs))
            E_input = Lambda(self.handle_input_E)([WT_output, inputs])
            min_coords = Lambda(self.handle_input_E, arguments={'get_min_coord': True})([WT_output, inputs])
            E_output = self.E(E_input)
            output = Lambda(self.update_array)([inputs, E_output, min_coords])
        except Exception as e:   
            print(f"Error", e)
        return output
    
```

```python
from keras import optimizers
def get_iou_vector(A, B):
    t = A>0
    p = B>0
    intersection = np.logical_and(t,p)
    union = np.logical_or(t,p)
    iou = (np.sum(intersection) + 1e-10 )/ (np.sum(union) + 1e-10)
    return iou

# def iou_metric(pred):
#     return tf.py_func(get_iou_vector, [pred>0.5], tf.float64)

def iou_metric(y_true, y_pred):
    y_pred = tf.cast(y_pred > 0.5, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) - intersection
    iou = (intersection + 1e-10) / (union + 1e-10)
    return iou

model.compile(optimizer=optimizers.Adam(lr=1e-3),
             loss=bce_dice_loss, metrics=['accuracy',iou_metric])
```

```python
# src: https://www.kaggle.com/aglotero/another-iou-metric
def get_iou_vector(A, B):
    t = A>0
    p = B>0
    intersection = np.logical_and(t,p)
    union = np.logical_or(t,p)
    iou = (np.sum(intersection) + 1e-10 )/ (np.sum(union) + 1e-10)
    return iou

def getIOUCurve(mask_org,predicted):
  thresholds = np.linspace(0, 1, 100)
  ious = np.array([get_iou_vector(mask_org, predicted > threshold) for threshold in thresholds])
  thres_best_index = np.argmax(ious[9:-10]) + 9
  iou_best = ious[thres_best_index]
  thres_best = thresholds[thres_best_index]
  return thresholds,ious,iou_best,thres_best
```

```python
def calculate_iou(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred)
    union = np.logical_or(y_true, y_pred)
    iou_score = np.sum(intersection) / np.sum(union)
    return iou_score
```
