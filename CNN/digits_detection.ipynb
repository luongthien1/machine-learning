#### Load data

```python
import keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train[0].shape
```

#### Scale

```python
import matplotlib.pyplot as plt

X_train = X_train / 255
X_test = X_test / 255
```

#### convert the 2-d matrix to a 1-d array

```python
X_train = X_train.reshape(-1,28,28,1)    #training set
X_test = X_test.reshape(-1,28,28,1)      #test set
```

```python
X_train.shape
```

#### Neuron Network

```python
from keras import datasets, layers, models
convolutional_neural_network = models.Sequential([
    layers.Conv2D(filters=25, kernel_size=(3, 3), activation='relu', input_shape=(28,28,1)), 
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
convolutional_neural_network.compile(optimizer="adam", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
convolutional_neural_network.fit(X_train, y_train, epochs=12)
```

```python
convolutional_neural_network.save("Digits_detection_keras")
```

```python
convolutional_neural_network = keras.models.load_model("Digits_detection_keras")
```
